### PROBLEM 1 (20 pts.; 50/50 for code & discussion) 
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(10 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(10 pts.)*


````
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, pi, zeros, exp
from numpy.fft import rfft, irfft, rfft2, irfft2
````

````
N = 1000

def sine_wave(n,N):
  return (sin(pi*n/N)*sin(20*pi*n/N))

y_val = []
n_val = []

for i in range(1000):
  y_val.append(sine_wave(i,N))
  n_val.append(i)

plt.plot(n_val,y_val)
plt.show()
````

````
coef = []

def dft(y):
  N = len(y)
  c = zeros(N//2+1,complex)
  for k in range(N//2+1):
    for n in range(N):
      c[k] += y_val[n]*exp(-2j*pi*k*n/N)
  return c

coef = np.abs(dft(y_val))
plt.plot(coef)
#add zoomed in plot on white space
````

A workable compromise is that if a value in the Fourier transform of the point spread function is smaller than a certain amount $\epsilon$ you don't divide by it-- just leave that coefficient alone.
````
blur = np.loadtxt(file_path+"blur.txt")

plt.imshow(blur)
plt.show()

def f(x,y):
  return exp(-((x)**2+(y)**2)/(2*sigma**2))

sigma = 25
N = 1024
````

````
PSF_array = np.empty([N,N], float)

for i in range(-N//2,N//2):
  for j in range(-N//2,N//2):
    PSF_array[i][j] = f(i,j)

plt.imshow(PSF_array)
````

````
def psf(x,y,h,k):
  return exp(-((x-h)**2+(y-k)**2)/(2*sigma**2))

psf_array = []

for i in range(len(blur)):
  row = []
  if i <= len(blur)/2:
    h = 0
  else:
    h = 1024
  for j in range(len(blur)):
    if j <= len(blur)/2:
      k = 0
    else:
      k = 1024
    row.append(psf(i,j,h,k))
  psf_array.append(row)

plt.imshow(psf_array)
````

````
blur_rfft = rfft2(blur)
psf_rfft = rfft2(psf_array)

unblur = np.empty(blur_rfft.shape, complex)

for i in range(len(blur_rfft)):
  for j in range(len(blur_rfft[0])):
    if np.abs(psf_rfft[i,j]) < 10**-3:
      unblur[i,j] = blur_rfft[i,j]
    else:
      unblur[i,j] = blur_rfft[i,j]/psf_rfft[i,j]

plt.imshow(irfft2(unblur))
````
